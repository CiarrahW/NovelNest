```python
import sqlite3
import pandas as pd
from pathlib import Path
```

```python
DB_PATH = Path("jinjiang_novels.db")

conn = sqlite3.connect(DB_PATH)
print(" 已连接到数据库：", DB_PATH)
```

```python
query_tables = "SELECT name FROM sqlite_master WHERE type='table';"
tables = pd.read_sql(query_tables, conn)
print(" 数据库中的所有表：")
tables
```

```python
table_name = "book"

df_info = pd.read_sql(f'PRAGMA table_info("{table_name}");', conn)
df_info
```

```python
table_name = "book"

df_preview = pd.read_sql(f'SELECT * FROM "{table_name}" LIMIT 20;', conn)
df_preview
```
